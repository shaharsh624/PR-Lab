### Preprocessing


```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, Binarizer
```

```python
df1 = pd.read_csv("../practical3//diabetes.csv")
```

```python
df2 = df1.drop("Outcome", axis=1)
df2
```

```python
min_max_scaler = MinMaxScaler(feature_range=(0, 1))
scaled_feat = min_max_scaler.fit_transform(df2)
df3 = pd.DataFrame(scaled_feat, columns=df2.columns)
df3
```

```python
standard_scaler = StandardScaler()
scaled_feat = standard_scaler.fit_transform(df2)
df3 = pd.DataFrame(scaled_feat, columns=df2.columns)
df3
```

```python
binarizer = Binarizer(threshold=30)
bin_feat = binarizer.fit_transform(df2)
df3 = pd.DataFrame(bin_feat, columns=df2.columns)
df3
```

### PCA


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("../practical4/Iris.csv")
df
```

```python
X = df.drop("Species", axis=1)
y = df["Species"]
```

```python
X
```

```python
y
```

```python
std_scaler = StandardScaler()
std_feat = std_scaler.fit_transform(X)
X_scaled = pd.DataFrame(std_feat, columns=X.columns)
X_scaled
```

```python
pca = PCA(n_components=2)
pca_feat = pca.fit_transform(X_scaled)
df4 = pd.DataFrame(pca_feat, columns=["PC1", "PC2"])
df4
```

```python
df5 = pd.concat([df4, y], axis=1)
df5
```

```python
colors = ["red", "green", "blue"]
species_names = ["Iris-setosa", "Iris-versicolor", "Iris-virginica"]
for sp, cl in zip(species_names, colors):
    class_sp = df5["Species"] == sp
    plt.scatter(df5.loc[class_sp, "PC1"], df5.loc[class_sp, "PC2"], c=cl, label=sp)
    plt.legend()
    plt.grid()
```

```python
df1 = pd.read_csv("../practical4//wine_data.csv")
df1
```

```python
X = df1.drop(["class_label"], axis=1)
y = df1["class_label"]
```

```python
scaler = StandardScaler() 
X_standardized = scaler.fit_transform(X) 
X_standardized_df = pd.DataFrame(X_standardized, columns=X.columns)
X_standardized_df
```

```python
plt.figure(figsize=(6, 6))
X_standardized_df.boxplot()
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
cov_matrix = X_standardized_df.cov()
cov_matrix
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```
